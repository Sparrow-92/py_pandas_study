### 다변량 그래프 그리기

3개 이상의 변수를 사용한 그래프는 '다변량 그래프'

지불 금액(total_bill)과 팁(tip)만을 이용한 산점도 그래프에 성별, 식사 비용이라는 변수가 추가될 경우 점의 생상이나 크기를 다르게 하는 방법으로 표현가능.

산점도 그래프에 성별을 새 변수로 추가.
성별은 색상으로 구분. 다만 Female, Male과 같은 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용될 수 없음. **(대신 정수사용!)**

여성은 0, 남성은 1로 치환하는 함수를 먼저 작성.

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

recode_sex 메서드가 반환한 값은 데이터프레임에 추가.
해당 값은 데이터프레임에 sex_color 열을 새로 생성하여 추가.
sex 열에 recode_sex 함수를 브로드캐스팅하려면 apply 메서드 사용.(10장에서 계속)

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

여기에 테이블당 인원 수 (size)를 산점도 그래프에 추가. (점의 크기로 표현)
scatter 메서드에 s,c 인자값으로 테이블당 인원 수와 성별의 치환값을 전달한 것은 아래와 같음.

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

# Chapter 4_3 seaborn 라이브러리 자유자재로 사용하기

주어진 데이터 외, seaborn 라이브러리 자체를 활용한 그래프 그리기.

### 히스토그램
단변량 그래프

```python
import seaborn as sns
tips=sns.load_dataset("tips")
```

seaborn 라이브러리로 히스토그램을 그리려면 subplots, distplot 메서드를 사용.
subplots 메서드로 기본 틀을 만들고 distplot 메서드에 total_bill 열 데이터를 전달하여 히스토그램을 그림.

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Hostogram with Density Plot')
```

distplot 메서드는 히스토그램과 밀집도 그래프를 같이 그림.
밀집도 그래프를 제외하고 싶다면 kde 인자 값을 False로 설정.

밀집도 그래프: 주어진 데이터를 정규화시켜 넓이가 1이 되도록 그린 그래프.

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

반대로, 밀집도 그래프만 나타내려면, hist 인자를 False로 지정.

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

실무환경에서는 효과적인 데이터 시각화를 위해 여러 그래프를 한번에 출력. 
아래는 히스토그램과 밀집도를 그리는 distplot 메서드에 rug 인자를 추가하여 양탄자 그래프를 추가한 것.

양탄자 그래프: 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프.

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

count 그래프는 이산값을 나타내는 그래프.
countplot 메서드에 tips 데이터프레임 중 day 열 데이터를 넣어 그린 그래프.

```python
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

### 다양한 종류의 이변량 그래프 그리기

1. 산점도 그래프
regplot 메서드를 이용하여 산점도 그래프와 회귀선을 함께 그림.
회귀선을 제거한 산점도 그래프는 fit_reg 인자를 False로 지정.

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Sacatterplot of Total Bill and Tip')
ax.set_xlabel('Total bill')
ax.set_yalbel('Tip')
```

이번에는 산점도 그래프와 히스토그램을 한 번에 그려주는 jointplot 메서드를 이용.
x, y인자에 원하는 열 이름을 지정하고 data 이자 값으로 데이터프레임을 지정하여 그래프를 그림.

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=3, y=1.03)
joint.show()
```

산점도 그래프는 점이 겹쳐 보인경우 점을 구분하기 어렵다는 단점이 있음.
보다 구분하기 쉬운 시각화 방법은 육각 그래프(hexbin)을 사용하는 것.
육각 그래프는 jointplot 메서드를 그대로 사용하는 대신 kind 인자값은 hex로 지정.

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

### 이차원 밀집도 그리기

이차원 밀집도는 kdeplot 메서드를 통해 만들 수 있음. (shade 인자 값을 True로 지정하면 그래프에 음영 효과를 줄 수 있음.)

```python
ax = plt.subplots()
ax = sns.kdeplot(x=tips['total_bill'],
                y=tips['tip'],
                shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

### 관계 그래프 그리기

과계 그래프는 지금까지 언급된 그래프를 종합한 그래프.
pairplot메서드에 데이터 프레임을 넣어 그릴 수 있음.

다만, 중복된 정보가 표현된다는 단점이 있으므로 그래프를 좌상향의 사선을 기준으로 위치를 직접 지정하여 중복된 그래프를 원하는 그래프 형태로 시각화 하여 해결할 수 있음.

map_upper 메서드는 대각선을 기준으로 위쪽에 그릴 그래프를,
map_lower 메서드는 대각선을 기준으로 아래쪽에 그릴 그래프를 각각 지정.
map_diag는 대각선을 중심으로 그래프를 그림. 

아래 관계 그래프는 이차원 밀집도를 대각선 기준으로 아래, 산점도 그래프는 대각선 기준 위쪽으로 그리고 히스토그램과 밀집도는 대각선을 중심으로 배치한 것.

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
plt.show()
```
